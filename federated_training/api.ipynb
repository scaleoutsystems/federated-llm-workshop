```python
from scaleout import Scaleout as APIClient
import time
import uuid
import json
import numpy as np
import collections
```

```python
client = APIClient(host="localhost", port=8092)
```

```python
client
```

```python
client.set_active_package('package.tgz', "numpyhelper", "package")
client.set_active_model('seed.npz')
seed_model = client.get_active_model()
print(seed_model)
```

```python
client.get_active_clients()
```

```python
session_config = {
                    "name": "My Session",
                    "helper": "numpyhelper",
                    "aggregator": "fedavg",
                    "model_id": seed_model['model'],
                    "rounds": 1,
                    "round_timeout": 60
                 }

result_fedavg = client.start_session(**session_config)
```
